## Lesson 1: Fibonacci & Base Cases (Easy)

### Core Concept
Base cases are the foundation of recursion. They stop the recursion and return a result directly without making further calls.

### Key Patterns
- **Identify base cases first**: What is the simplest version of the problem?
- **Recursive case**: How do I break the problem into smaller subproblems?
- **Trust the recursion**: Assume fib(n-1) and fib(n-2) work correctly.

### Example: Fibonacci
```python
def fib(n):
    # Base cases - define the smallest problems you can solve directly
    if n == 0: return 0
    if n == 1: return 1
    # Recursive case - break problem into smaller pieces
    return fib(n-1) + fib(n-2)
```

### What to Analyze
1. **Recursion tree**: Draw out calls for small n (e.g., n=4) to visualize overlap
2. **Base case redundancy**: How many times is fib(1) or fib(0) recalculated?
3. **Time complexity**: O(2^n) - exponential due to overlapping subproblems

### Common Mistakes
- Forgetting one base case
- Infinite recursion (recursive case doesn't move toward base case)
- Not testing with edge cases (n=0, n=1)

### After solving Q1-5, you should understand:
- How to write recursive functions from problem definition
- The importance of base cases
- How to trace recursion execution
- Why naive recursion can be inefficient

## Lesson 2: Recursion on Strings/Arrays (Easy → Medium)

### Core Concept
Recursion on linear structures (arrays/strings) typically shrinks the problem by 1 element per call. Common strategies: index-based recursion, or removing an element.

### Key Patterns
- **Index recursion**: `solve(arr, index)` where index moves toward boundary
- **Shrinking input**: `solve(arr[1:])` processes smaller array each call
- **Accumulator pattern**: `solve(arr, index, accumulator)` builds result during recursion

### Example: Recursive Sum
```python
def recursive_sum(arr, index=0):
    # Base case: when we've processed all elements
    if index == len(arr): return 0
    # Recursive case: current element + sum of rest
    return arr[index] + recursive_sum(arr, index + 1)

# Or with accumulator:
def recursive_sum_acc(arr, index, acc=0):
    if index == len(arr): return acc
    return recursive_sum_acc(arr, index + 1, acc + arr[index])
```

### Example: Palindrome Check
```python
def is_palindrome(s):
    if len(s) <= 1: return True
    if s[0] != s[-1]: return False
    return is_palindrome(s[1:-1])
```

### What to Analyze
1. **How to shrink the problem**: Which element(s) to remove/skip?
2. **Base case**: When is the problem solved?
3. **Combining results**: How do recursive results combine?

## Lesson 3: Binary Recursion Patterns (Medium)

### Core Concept
Binary recursion makes 2+ recursive calls per function call. This creates exponential branching unless pruned or memoized.

### Key Patterns
- **Include/Exclude**: For each element, solve with it AND without it
- **Divide-and-Conquer**: Split problem in half, solve independently, combine
- **Decision tree**: Each call makes a binary choice that branches execution path

### Example: Generate Subsequences (Include/Exclude)
```python
def subsequences(s, index=0, current="", result=None):
    if result is None: result = []
    if index == len(s):
        result.append(current)
        return result
    # Decision 1: Include current character
    subsequences(s, index + 1, current + s[index], result)
    # Decision 2: Exclude current character
    subsequences(s, index + 1, current, result)
    return result
```

### Example: Power Function (Divide-and-Conquer)
```python
def pow(x, n):
    if n == 0: return 1
    if n < 0: return 1 / pow(x, -n)
    # Optimize: compute half power, square it
    half = pow(x, n // 2)
    if n % 2 == 0:
        return half * half
    else:
        return half * half * x
```

### What to Analyze
1. **Decision points**: What choices are made at each level?
2. **Branching factor**: How many recursive calls per invocation?
3. **Pruning**: Can we eliminate branches that cannot lead to valid solutions?

## Lesson 4: Backtracking Fundamentals (Medium)

### Core Concept
Backtracking = binary recursion + choose-explore-unchoose pattern. Build solutions incrementally; undo choices that don't lead to valid solutions.

### Core Pattern: Choose-Explore-Unchoose
```python
def backtrack(candidates, current_solution, result):
    if is_solution(current_solution):
        result.append(current_solution.copy())
        return
    for choice in get_choices(candidates):
        # 1. CHOOSE: Add candidate to current solution
        current_solution.append(choice)
        # 2. EXPLORE: Recursively build solution
        backtrack(remaining, current_solution, result)
        # 3. UNCHOOSE: Remove candidate (restore state)
        current_solution.pop()
```

### Example: Subsets
```python
def subsets(nums):
    result = []
    def backtrack(start, current):
        result.append(current.copy())
        for i in range(start, len(nums)):
            current.append(nums[i])
            backtrack(i + 1, current)
            current.pop()  # Undo choice
    backtrack(0, [])
    return result
```

### Key Differences from Include/Exclude
- **Include/Exclude**: Questions all elements in order (binary decision per element)
- **Backtracking**: Choose from **remaining** candidates, **undo** after exploring
- **Result**: Avoids duplicate solutions and enables pruning

### What to Analyze
1. **State variable**: What must be restored when we unchoose?
2. **Pruning opportunity**: When should we skip a branch?
3. **Solution check**: When is a complete solution found?

- Choose-explore-unchoose lifecycle
- Why state restoration matters
- How to generate all permutations/combinations without duplicates

## Lesson 5: Constraint Backtracking (Medium → Hard)

### Core Concept
Add **constraints** to backtracking: prune branches early when constraints are violated. Reduces search space dramatically.

### Key Patterns
- **Target value**: Prune when partial sum exceeds target
- **Valid building blocks**: Only consider remaining candidates
- **Early termination**: Stop exploring a branch if impossible to reach goal

### Example: Combination Sum (Unlimited Reuse)
```python
def combination_sum(candidates, target):
    result = []
    def backtrack(start, current, remaining):
        if remaining == 0:
            result.append(current.copy())
            return
        if remaining < 0:  # Pruning: exceeded target
            return
        for i in range(start, len(candidates)):
            current.append(candidates[i])
            # Reuse candidates[i] by passing start=i again
            backtrack(i, current, remaining - candidates[i])
            current.pop()
    backtrack(0, [], target)
    return result
```

### Example: Generate Parentheses (Constraint: Valid Pairs)
```python
def generate_parentheses(n):
    result = []
    def backtrack(current, open_count, close_count):
        if len(current) == 2 * n:
            result.append(current)
            return
        if open_count < n:  # Can add more opens
            backtrack(current + "(", open_count + 1, close_count)
        if close_count < open_count:  # Can add closes (valid constraint)
            backtrack(current + ")", open_count, close_count + 1)
    backtrack("", 0, 0)
    return result
```

### What to Analyze
1. **Pruning condition**: What makes a branch unsolvable?
2. **Constraint checking**: When do we stop exploring?
3. **Early returns**: How much branching can we avoid?

## Lesson 6: Grid & Board Backtracking (Hard)

### Core Concept
2D grid adds **spatial state**: visited cells, boundary checks, 4-directional movement. Backtrack by marking/unmarking visited cells.

### Key Patterns
- **Visited set**: Track which cells we've used in current path
- **4-directional exploration**: Up, down, left, right
- **Boundary checks**: Stay within grid bounds
- **State restoration**: Unmark visited when backtracking

### Example: Word Search (Path Finding)
```python
def word_search(board, word):
    def backtrack(r, c, index, visited):
        if index == len(word):  # Found entire word
            return True
        # Try all 4 directions
        for dr, dc in [(0,1), (0,-1), (1,0), (-1,0)]:
            nr, nc = r + dr, c + dc
            if (0 <= nr < len(board) and 0 <= nc < len(board[0]) and
                (nr, nc) not in visited and board[nr][nc] == word[index]):
                visited.add((nr, nc))
                if backtrack(nr, nc, index + 1, visited):
                    return True
                visited.remove((nr, nc))  # Undo explore
        return False
    visited = set()
    for r in range(len(board)):
        for c in range(len(board[0])):
            if board[r][c] == word[0]:
                visited.add((r, c))
                if backtrack(r, c, 1, visited):
                    return True
                visited.remove((r, c))
    return False
```

### Example: N-Queens (Constraint Placement)
```python
def solve_nqueens(n):
    result = []
    board = [["." for _ in range(n)] for _ in range(n)]
    cols, diag1, diag2 = set(), set(), set()  # Track occupied lines
    def backtrack(row):
        if row == n:
            result.append(["".join(row) for row in board])
            return
        for col in range(n):
            if col in cols or row-col in diag1 or row+col in diag2:
                continue  # Pruning: conflict detected
            board[row][col] = "Q"
            cols.add(col); diag1.add(row-col); diag2.add(row+col)
            backtrack(row + 1)
            board[row][col] = "."
            cols.remove(col); diag1.remove(row-col); diag2.remove(row+col)
    backtrack(0)
    return result
```

## Lesson 7: Advanced Pruning + Recursion Depth (Hard)

### Core Concept
Maximize pruning through: sorting for better bounds, mathematical constraints, failed branch memoization, and depth limits.

### Key Patterns
- **Sorting**: Order candidates so invalid branches fail early
- **Bounds checking**: Mathematical inequalities that eliminate branches
- **Memoization**: Cache subproblems to avoid recomputation
- **Backtracking + DP**: Combine techniques

### Example: Matchsticks to Square (Smart Pruning)
```python
def makesquare(matchsticks):
    if len(matchsticks) < 4 or sum(matchsticks) % 4 != 0:
        return False
    matchsticks.sort(reverse=True)  # Sort descending for faster pruning
    target = sum(matchsticks) // 4
    if matchsticks[0] > target:  # Prune: single stick > target
        return False
    sides = [0] * 4
    def backtrack(index):
        if index == len(matchsticks):
            return all(s == target for s in sides)
        for i in range(4):
            if sides[i] + matchsticks[index] <= target:
                sides[i] += matchsticks[index]
                if backtrack(index + 1):
                    return True
                sides[i] -= matchsticks[index]
                if sides[i] == 0:  # Prune: symmetry
                    break
        return False
    return backtrack(0)
```

### Example: Partition to K Equal Sum (Accumulator Pruning)
```python
def canPartitionKSubsets(nums, k):
    if sum(nums) % k != 0: return False
    target = sum(nums) // k
    nums.sort(reverse=True)
    buckets = [0] * k
    def backtrack(index):
        if index == len(nums):
            return all(b == target for b in buckets)
        for i in range(k):
            if buckets[i] + nums[index] <= target:
                buckets[i] += nums[index]
                if backtrack(index + 1):
                    return True
                buckets[i] -= nums[index]
                if buckets[i] == 0: break  # Prune: symmetry
        return False
    return backtrack(0)
```

### Example: 24 Game (Multi-strategy)
```python
def judgePoint24(cards):
    def backtrack(nums):
        if len(nums) == 1:
            return abs(nums[0] - 24) < 1e-6
        for i in range(len(nums)):
            for j in range(len(nums)):
                if i != j:
                    a, b = nums[i], nums[j]
                    remaining = [nums[k] for k in range(len(nums)) if k != i and k != j]
                    # Try all operations
                    for op_result in [a + b, a - b, a * b]:
                        if backtrack(remaining + [op_result]):
                            return True
                    if b != 0 and backtrack(remaining + [a / b]):
                        return True
        return False
    return backtrack(list(map(float, cards)))
```

### Example: Regular Expression Matching (DP + Recursion)
```python
def isMatch(s, p):
    memo = {}
    def dp(s_idx, p_idx):
        if (s_idx, p_idx) in memo:
            return memo[(s_idx, p_idx)]
        if p_idx == len(p):
            return s_idx == len(s)
        first_match = s_idx < len(s) and (p[p_idx] == "." or p[p_idx] == s[s_idx])
        if p_idx + 1 < len(p) and p[p_idx + 1] == "*":
            # Option 1: Skip this char-star in pattern
            # Option 2: Match and move in string, stay in pattern
            result = dp(s_idx, p_idx + 2) or (first_match and dp(s_idx + 1, p_idx))
        else:
            result = first_match and dp(s_idx + 1, p_idx + 1)
        memo[(s_idx, p_idx)] = result
        return result
    return dp(0, 0)
```
